#### Step - 3 Check if the data is balanced or not. ..

```python
print(df['label'].value_counts())
```

```python
print(df['label'].value_counts(normalize=True)*100)
```

```python
df['review_length'] = df['text'].apply(len)
```

```python
print(f"Average Length: {df['review_length'].mean()}")
print(f"Shortest Review: {df['review_length'].min()}")
print(f"Longest Review: {df['review_length'].max()}")
```

```python
# let's see the shortest and the longest review 
print(df.loc[df['review_length'].idxmin(), 'text'])
print(df.loc[df['review_length'].idxmax(), 'text'][:1000])
```

```python
# Now check the positive and negative reviews !!!
print(df[df['label'] == 0]['text'].iloc[0][:300])
print(df[df['label'] == 1]['text'].iloc[0][:300])
```
